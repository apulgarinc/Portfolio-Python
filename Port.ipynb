```python
#import librarias

import pandas as pd
import numpy as np
import requests
import datetime
import plotly
import cufflinks as cf

```

```python
def get_cotacao(ticker='^BVSP', source='yahoo', b_date='01/01/2000', e_date=datetime.datetime.strptime(str(datetime.date.today()), "%Y-%m-%d").strftime("%m-%d-%Y")):
    import pandas as pd
    from pandas_datareader import data as get
    return get.DataReader(ticker, data_source=source, start=b_date, end=e_date)
def justclose(data,name_of_ticker):
    data.drop(data[['High','Low','Open','Close','Volume']],axis=1, inplace=True)
    data.rename(columns={'Adj Close':name_of_ticker},inplace=True)
    data.dropna(inplace=True)
def base(b,asset):
    b=b.join(asset,how='outer')
    b.dropna(inplace=True)
    b.rename(index={})
    return b
def get_close_all_assets(list_of_assets):
    data=pd.DataFrame(index={'Date'})
    for i in list_of_assets:
        asset=get_cotacao(i)
        justclose(asset,i)
        data=base(data,asset)
    data.index.names = ['Date']
    return data
    
    

    
```

```python
list_tickets=['RADL3.SA','OIBR4.SA','LAME4.SA','UGPA3.SA','EQTL3.SA','BBSE3.SA','LREN3.SA','ENBR3.SA','JBSS3.SA','KLBN11.SA','BBDC3.SA','BBDC4.SA','BRKM5.SA','BBAS3.SA','CMIG4.SA','BRML3.SA','ITSA4.SA','ECOR3.SA','MULT3.SA','CCRO3.SA','SBSP3.SA','PETR3.SA','USIM5.SA']
```

```python
data=get_close_all_assets(list_tickets)
```

```python
data
```

```python
#optimize the protfolio
from pypfopt.efficient_frontier import EfficientFrontier as efront
from pypfopt import risk_models
from pypfopt import expected_returns
```

```python
#calcular el valor anualizado de los retornos y la matriz de de covarianza anualzada para los retornos
mu=expected_returns.mean_historical_return(data)
S=risk_models.sample_cov(data)

```

```python
import plotly.express as px

df = S
fig = px.imshow(df)
fig.show()
```

```python
#optimizar maximo sharpe ratio
ef=efront(mu,S)
weights=ef.max_sharpe()
cleaned_weights=ef.clean_weights()
ef.portfolio_performance(verbose=True)
```

```python
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices
portfolio_val = 5000
lastest_prices= get_latest_prices(data)
weights =cleaned_weights
da=DiscreteAllocation(weights, lastest_prices, total_portfolio_value= portfolio_val)
allocation, leftover = da.lp_portfolio()
a=pd.DataFrame(lastest_prices)
```

```python
a.columns=['Preco']
a['Ticker']=a.index
a.reset_index(drop=True, inplace=True)
a
```

```python
print('Alocacao: ', allocation)
print('Reais que faltam: $',leftover)
```

```python
portafolio=pd.DataFrame()
portafolio['Ticker']=allocation.keys()
portafolio['Quantidade de acoes com '+str(portfolio_val)+' Reais']=allocation.values()
portafolio=portafolio.merge(a,how='inner', on='Ticker')
portafolio
```

```python
cf.set_config_file(theme='pearl',sharing='public',offline=True)
data[allocation.keys()].iplot(y=allocation.keys(),
               width=2.0,
               xTitle="Date", yTitle="Adj Close", title="Alejandro's portfolio")
```
